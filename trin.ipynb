```python
import psutil
import platform

def get_system_info():
    """Mengambil informasi spesifikasi sistem dan penggunaan sumber daya."""

    print("Informasi Sistem:")
    print(f"  Sistem Operasi: {platform.system()} {platform.release()} ({platform.architecture()[0]})")
    print(f"  Nama Node: {platform.node()}")
    print(f"  Versi Python: {platform.python_version()}")

    print("\nInformasi CPU:")
    print(f"  Nama CPU: {platform.processor()}")
    print(f"  Jumlah Core Fisik: {psutil.cpu_count(logical=False)}")
    print(f"  Jumlah Core Logis: {psutil.cpu_count(logical=True)}")
    print(f"  Penggunaan CPU (%): {psutil.cpu_percent(interval=1)}")

    print("\nInformasi Memori (RAM):")
    mem = psutil.virtual_memory()
    print(f"  Total: {mem.total / (1024**3):.2f} GB")
    print(f"  Tersedia: {mem.available / (1024**3):.2f} GB")
    print(f"  Digunakan: {mem.used / (1024**3):.2f} GB")
    print(f"  Persentase Penggunaan: {mem.percent}%")

    print("\nInformasi Disk:")
    partitions = psutil.disk_partitions()
    for partition in partitions:
        print(f"  Partisi: {partition.device}")
        try:
            usage = psutil.disk_usage(partition.mountpoint)
            print(f"    Total: {usage.total / (1024**3):.2f} GB")
            print(f"    Digunakan: {usage.used / (1024**3):.2f} GB")
            print(f"    Bebas: {usage.free / (1024**3):.2f} GB")
            print(f"    Persentase Penggunaan: {usage.percent}%")
        except PermissionError:
            print("    Akses ditolak.")

    print("\nInformasi Jaringan:")
    interfaces = psutil.net_if_addrs()
    for interface_name, addresses in interfaces.items():
        print(f"  Antarmuka: {interface_name}")
        for address in addresses:
            if str(address.family) == 'AddressFamily.AF_INET':
                print(f"    Alamat IP: {address.address}")
                print(f"    Netmask: {address.netmask}")
                print(f"    Broadcast: {address.broadcast}")
            elif str(address.family) == 'AddressFamily.AF_PACKET':
                print(f"    Alamat MAC: {address.address}")
                print(f"    Netmask: {address.netmask}")
                print(f"    Broadcast: {address.broadcast}")

if __name__ == "__main__":
    get_system_info()
```

```python
import psutil
import platform
import GPUtil

def get_cpu_gpu_info():
    """Mendapatkan informasi CPU dan GPU."""

    # Informasi CPU
    print("Informasi CPU:")
    print(f"  Jumlah Core Fisik: {psutil.cpu_count(logical=False)}")
    print(f"  Jumlah Thread (Core Logis): {psutil.cpu_count(logical=True)}")
    print(f"  Kecepatan Clock CPU (MHz): {psutil.cpu_freq().current:.2f}")

    # Informasi GPU
    print("\nInformasi GPU:")
    try:
        gpus = GPUtil.getGPUs()
        for gpu in gpus:
            print(f"  Nama GPU: {gpu.name}")
            print(f"  Memori GPU (MB): {gpu.memoryTotal}")
    except ModuleNotFoundError:
        print("  Pustaka GPUtil tidak ditemukan. Jalankan 'pip install GPUtil' untuk menginstalnya.")
    except Exception as e:
        print(f"  Terjadi kesalahan saat mengambil informasi GPU: {e}")

if __name__ == "__main__":
    get_cpu_gpu_info()
```
